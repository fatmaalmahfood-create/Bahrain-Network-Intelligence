```python
features.drop(columns= 'speed_ratio', inplace=True)
```

```python
features = features.sort_values(['area', 'year', 'quarter'])
```

```python
features.head(16)
```

```python
features['area_rolling_d_mbps'] = (features.groupby('area')['avg_d_mbps'].transform(lambda x: x.shift(1)
    .rolling(2, min_periods=1).mean()).fillna(features['avg_d_mbps'].mean()))
```

```python
features.drop(columns= 'city_rolling_d_mbps', inplace=True)
```

```python
# Year trend calculation for the model to know what year specifically the record is token from, fro more effecient detection of network improvements over the years
```

```python
features['year_trend'] = features['year'] - features['year'].min()
```
