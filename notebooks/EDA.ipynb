```python
import geopandas as gpd
```

```python
towers
```

```python
gdf = gpd.GeoDataFrame(
    towers,
    geometry=gpd.points_from_xy(towers['Longitude'], towers['Latitude']),
    crs="EPSG:4326"
)
```

```python
gdf.dtypes
```

```python
towers.dtypes
```

```python
gdf.plot(
    cmap="viridis",
    legend=True,
    figsize=(8,6)
)

plt.title("towers distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt

gdf.plot(figsize=(8, 8), markersize=5)

plt.title("Tower Locations in Bahrain")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
gdf.plot(
    column='Operator',
    categorical=True,
    legend=True,
    figsize=(8, 8),
    markersize=10
)

plt.title("Tower Distribution by Operator")
plt.savefig('assets/figures/towers_dist_by_operators.png', dpi=150)
plt.show()
```

```python
gdf.plot(
    column='RAT Type',
    categorical=True,
    legend=True,
    figsize=(8, 8),
    markersize=10
)

plt.title("Tower Distribution by RAT Type (LTE vs 5G)")
plt.savefig('assets/figures/towers_dist_by_RAT_Type.png', dpi=150)
plt.show()
```

```python
gdf_active = gdf[gdf['Visible'] == True]
```
